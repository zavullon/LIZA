```python
def export_kaggle(df_test, proba_test, save=True, path_save="my_submission_vanilla_model.csv"):
    """
    Export submissions with good kaggle format.
    df_test : (pandas dataframe) test set
    proba_test : (numpy ndarray) probabilities as numpy ndarray you get using method .predict_proba()[:,1]
    save : (bool) if set to True, it will save csv submission in path_save path.
    path_save : (str) path where to save submission.
    return : dataframe for submission
    """
    proba_serie = pd.Series(proba_test, index=df_test.index)
    df_submit = pd.concat([df_test["id"], proba_serie], axis=1)
    df_submit.columns = ["id", "decision"]
    df_submit.to_csv(path_save, index=False)
    return df_submit

def check_test(result, expected, display_error):
    """
    Testing your results.
    """
    if result == expected:
        print("1 test passed.")
    else:
        print(display_error)
```

#### Creation of features
From f29 (log of total assets), I want to get the level of total assets (f65). And then I multiply f65 to other columns that has total assets as denominator to get different financial numbers like "net profit" (f1), "total liabilities" (f2), etc.
$$
$$
With "total liabilities" (f65f2), I other columns that have "total liabilities" as denominator to get more new features like  book value of equity (f8), operating expenses (f34).


```python
# Create a new column representing total assets
def createcols(df_train0,df_test0):
    df_train1=df_train0.copy()
    df_test1=df_test0.copy()
    features_cols = [col for col in df_train1.columns if "f" in col]
    #listf=["f1","f2","f3","f9","f10","f24","f51"]
    listf=["f1","f2","f3","f9","f10","f22","f24","f35","f36","f38","f48","f51"]
    idx=df_train1.columns.get_loc(features_cols[len(features_cols)-1])+1
    asset_train=np.exp(df_train1['f29'])
    asset_test=np.exp(df_test1['f29'])
    df_train1.insert(loc=idx, column='f65', value=asset_train)
    df_test1.insert(loc=idx, column='f65', value=asset_test)
    idcol=65
    for i in range(len(listf)):
        idx=idx+1
        idcol=idcol+1
        new_col_train=asset_train*df_train1[listf[i]]
        new_col_test=asset_test*df_test1[listf[i]]
        df_train1.insert(loc=idx, column="f65"+listf[i], value=new_col_train)
        df_test1.insert(loc=idx, column="f65"+listf[i], value=new_col_test)
        
#         plt.figure()
#         plt.hist((df_train1["f65"+listf[i]].dropna()),bins=50)
#         plt.title("histogram of "+"f65"+listf[i])
#         plt.show()
#     plt.figure()
#     plt.hist(np.log(df_train1[df_train1["f65f51"]!=0.0]["f65f51"].dropna()),bins=50)
#     plt.title("histogram of f65f51")
#     plt.show()
#     plt.figure()
#     plt.hist(np.log(df_train1["f65f2"].dropna()+0.001),bins=50)
#     plt.title("histogram of f65f2")
#     plt.show()
    
    df_train1.iloc[:,df_train1.columns.get_loc("f65f51")]=df_train1["f65f51"].apply(lambda x: np.log(x) if x>0 else np.nan)
    df_train1.iloc[:,df_train1.columns.get_loc("f65f2")]=df_train1["f65f2"].apply(lambda x: np.log(x) if x>0 else np.nan)
    
    df_test1.iloc[:,df_train1.columns.get_loc("f65f51")]=df_test1["f65f51"].apply(lambda x: np.log(x) if x>0 else np.nan)
    df_test1.iloc[:,df_train1.columns.get_loc("f65f2")]=df_test1["f65f2"].apply(lambda x: np.log(x) if x>0 else np.nan)
    
#     df_train1.iloc[:,df_train1.columns.get_loc("f65f36")]=df_train1["f65f36"].apply(lambda x: np.log(x) if x>0 else np.nan)
#     df_test1.iloc[:,df_train1.columns.get_loc("f65f36")]=df_test1["f65f36"].apply(lambda x: np.log(x) if x>0 else np.nan)
    
#     ll=["f51","f2","f36"]
#     for i in range(len(ll)):
#         plt.figure()
#         plt.hist((df_train1["f65"+ll[i]].dropna()),bins=50)
#         plt.title("histogram of "+"f65"+ll[i])
#         plt.show()
    listf1=["f8","f17","f26","f34","f50"]
    features_cols = [col for col in df_train1.columns if "f" in col]
    l_train=df_train1["f65f2"]
    l_test=df_test1["f65f2"]
    idx=df_train1.columns.get_loc(features_cols[len(features_cols)-1])
    for i in range(len(listf1)):
        idx=idx+1
        
        new_col_train=l_train*df_train1[listf1[i]]
        new_col_test=l_test*df_test1[listf1[i]]
        df_train1.insert(loc=idx, column="f65f2"+listf1[i], value=new_col_train)
        df_test1.insert(loc=idx, column="f65f2"+listf1[i], value=new_col_test)
    
    idx=idx+1
    new_col_train=df_train1["f65f10"]*df_train1["f59"]
    new_col_test=df_test1["f65f10"]*df_test1["f59"]
    df_train1.insert(loc=idx, column="f65f10f59", value=new_col_train)
    df_test1.insert(loc=idx, column="f65f10f59", value=new_col_test)
    
#     df_train1=df_train1.drop(listf,axis=1)
#     df_test1=df_test1.drop(listf,axis=1)
    return df_train1,df_test1

ooo=createcols(df_train0,df_test0)[0]
#ooo["f65f2"].isnull().sum()
ooo
```

## Creating linear combination of features
I realized that tree-based methods (Decision Tree, Random Forest, Gradient Boosting,etc) do not need much feature engineering like transformation to $x^2, x^3, ln(x)$, or standardization, outlier cleaning. They are not very sensitive to these data transformations.
$$
$$
However, they cannot study linear combinations of features (https://zhuanlan.zhihu.com/p/26444240). Therefore, I want to create more features using "+" or "-" between columns to see if we get useful things.
$$
$$
As there are too many new columns (more than 1800), I select the ones that are most correlated to "label" column of train dataset (although it might not be very accurate as "label" is only 0 and 1). I use the Spearman correlation as it is more appropriate if we use tree-based estimator.(http://gplearn.readthedocs.io/en/stable/intro.html#transformer).  I also found that it is possible to create features using SymbolicTransformer in gplearn (generic programming), but it is too complicated to use.


```python
def LinearCombinations(d_train,d_test):
    from scipy import stats
    df_train1=d_train.copy()
    df_test1=d_test.copy()
    features_cols = [col for col in df_train1.columns if "f" in col]
#     features_cols=features_cols[:19]
    coefs=[]
#     numpy.corrcoef
# scipy.stats.spearmanr
    
    df=pd.DataFrame(df_train1.iloc[:,:1])
    df_t=pd.DataFrame(df_test1.iloc[:,:1])
    idx=1
#     for i,col1 in enumerate(features_cols[:4]):
#         for j,col2 in enumerate(features_cols[i+1:6]):
    for i,col1 in enumerate(features_cols[:-1]):
        for j,col2 in enumerate(features_cols[i+1:]):
        # ***************************
        # One column minus another
        # ***************************
            new_col=df_train1[col1]-df_train1[col2]
            
            # Calculate the Spearman's rank correlation coefficient ()
            # ******************************************************
            cor=stats.spearmanr(new_col,d_train['label'])
            if cor[1]<0.05:
                df.insert(loc=idx, column=col1+'-'+col2, value=new_col)
                df_t.insert(loc=idx, column=col1+'-'+col2, value=df_test1[col1]-df_test1[col2])
                idx=idx+1
            coefs.append([col1+'-'+col2,cor[1]])
#     print(sorted(coefs))
    
    
    ccc=pd.DataFrame(coefs,columns=['col','coef'])
    ccc=ccc.sort_values(by='coef')
    ccc=ccc.iloc[:40,:]
    
    df=df.loc[:,ccc['col'].tolist()]    
    df_t=df_t.loc[:,ccc['col'].tolist()]  
#     df.insert(loc=0,column='id',value=d_train['id'])
#     df['label']=df_train1['label']
    print(df.columns)
    return df,df_t
        
df,df_t=LinearCombinations(df_train, df_test)
```

### XGBoost
Here we try the XGBoost library. It is more efficient than the Gradient Boosting method in sklearn, but the improvement in performance is not much. I did some GridSearchCV test to find the best parameters

#### Parameters tuning

```python
from xgboost.sklearn import XGBClassifier
def modelfit(alg, dtrain, predictors,useTrainCV=True, cv_folds=5, early_stopping_rounds=50):
    target='label'
    if useTrainCV:
        xgb_param = alg.get_xgb_params()
        xgtrain = xgb.DMatrix(dtrain[predictors].values, label=dtrain[target].values)
        cvresult = xgb.cv(xgb_param, xgtrain, num_boost_round=alg.get_params()['n_estimators'], nfold=cv_folds,
            metrics='auc', early_stopping_rounds=early_stopping_rounds)#, show_progress=False)
        alg.set_params(n_estimators=cvresult.shape[0])

    #Fit the algorithm on the data
    alg.fit(dtrain[predictors], dtrain[target],eval_metric='auc')

    #Predict training set:
    dtrain_predictions = alg.predict(dtrain[predictors])
    dtrain_predprob = alg.predict_proba(dtrain[predictors])[:,1]

    #Print model report:
    print("\nModel Report")
    print("Accuracy : %.4g" % metrics.accuracy_score(dtrain[target].values, dtrain_predictions))
    print("AUC Score (Train): %f" % metrics.roc_auc_score(dtrain[target], dtrain_predprob))
    plt.figure(figsize=(15,4))
    feat_imp = pd.Series(alg.booster().get_fscore()).sort_values(ascending=False)
    feat_imp.plot(kind='bar', title='Feature Importances')
    plt.ylabel('Feature Importance Score')
    plt.show()
    return alg
```

```python
def depthtest(data):
    features_cols = [col for col in data.columns if "f" in col]
    from sklearn.model_selection import GridSearchCV

#     param_test1 = { 'max_depth':range(3,10,2),'min_child_weight':range(1,6,2)}
    param_test1 = { 'max_depth':[2,3,4,5],'min_child_weight':[1,2,3]}
    # We first made a broad test of range(3,10,2) for max_depth, and range(1,6,2) for min_child_weight, then we adjust to be more precise
    gsearch1 = GridSearchCV(estimator = XGBClassifier(learning_rate =0.1, n_estimators=200,
                                                       gamma=0, subsample=0.9,colsample_bytree=0.6,
                                                      objective= 'binary:logistic', nthread=4,     scale_pos_weight=1, seed=27), 
                            param_grid = param_test1,     scoring='roc_auc',n_jobs=4,iid=False)
    gsearch1.fit(data.loc[:,features_cols],data['label'])
    return gsearch1.grid_scores_, gsearch1.best_params_,     gsearch1.best_score_
depthtest(df_tr)
```

```python
def gamma(data):
    
    features_cols = [col for col in data.columns if "f" in col]
    param_test3 = { 'gamma':[i/10.0 for i in range(0,6)]}
    gsearch1 = GridSearchCV(estimator = XGBClassifier(learning_rate =0.1, n_estimators=200,max_depth=3,
                                                       min_child_weight=1, subsample=0.9,colsample_bytree=0.6,
                                                      objective= 'binary:logistic', nthread=4,     scale_pos_weight=1, seed=27), 
                            param_grid = param_test3,     scoring='roc_auc',n_jobs=4,iid=False)
    gsearch1.fit(data.loc[:,features_cols],data['label'])
    return gsearch1.grid_scores_, gsearch1.best_params_,     gsearch1.best_score_
gamma(df_tr)
```

```python
def subsample(data):
    param_test = { 'subsample':[i/10.0 for i in range(6,10)], 'colsample_bytree':[i/10.0 for i in range(4,9)]}
    features_cols = [col for col in data.columns if "f" in col]
    gsearch1 = GridSearchCV(estimator = XGBClassifier(learning_rate =0.1, n_estimators=200,max_depth=3,
                                                       min_child_weight=1, gamma=0,
                                                      objective= 'binary:logistic', nthread=4,     scale_pos_weight=1, seed=27), 
                            param_grid = param_test,     scoring='roc_auc',n_jobs=4,iid=False)
    gsearch1.fit(data.loc[:,features_cols],data['label'])
    return gsearch1.grid_scores_, gsearch1.best_params_,     gsearch1.best_score_
subsample(df_tr)
```

```python
def search_n_estimators(df_tr):
    from sklearn.grid_search import GridSearchCV
    param_test2 = { 'n_estimators':[300,400]}
    gsearch2 = GridSearchCV(estimator = ensemble.GradientBoostingClassifier(max_depth=4, learning_rate=0.05, subsample=0.9), param_grid = param_test2, scoring='roc_auc',iid=False, cv=5)
    gsearch2.fit(df_tr.loc[:,features_cols],df_tr['label'])
    return gsearch2.grid_scores_, gsearch2.best_params_, gsearch2.best_score_
search_n_estimators(df_tr)    
```

```python
def search_learning_rate(df_tr):
    from sklearn.grid_search import GridSearchCV
    param_test2 = { 'learning_rate':np.linspace(0.05,0.08,4)}
    gsearch2 = GridSearchCV(estimator = ensemble.GradientBoostingClassifier(n_estimators=400,max_depth=4, subsample=0.9), param_grid = param_test2, scoring='roc_auc',iid=False, cv=5)
    gsearch2.fit(df_tr.loc[:,features_cols],df_tr['label'])
    return gsearch2.grid_scores_, gsearch2.best_params_, gsearch2.best_score_

search_learning_rate(df_tr)
```
